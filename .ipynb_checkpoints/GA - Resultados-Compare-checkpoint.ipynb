### Fitness

```python
prob_mutation='0.001 0.01 0.05 0.1 0.2 0.5 0.9'.split()
prob_crossover='1 0.9'.split()
init_population='1000'.split()
generations='1500'
```

```python
for p in init_population:
    for c in prob_crossover:
        print("prob_crossover: %s" % c)
        f, ax = plt.subplots(figsize=(10, 5))
        for m in prob_mutation:        
            print("init_population: %s" % init_population)
            try:
                df = pd.read_csv(path+'input_dataset_go_{}_{}_{}_{}_norm'.format(m, c, p, generations)+'/avg_fitness.csv')
            except:
                continue
            plt.plot(df.avg_fitness[10:-1], label='{}'.format(m))
        plt.title('Fitness do melhor resultado a cada Iterao', loc='left', y=1.04 )
        plt.xlabel('Geraes')
        plt.ylabel('Fitness')

        # Now add the legend with some customizations.
        legend = ax.legend(loc='upper right', shadow=True)

        plt.show()    
        print(list(df.avg_fitness)[0], " ", list(df.avg_fitness)[-1])
```

```python
def plot_count_clients(df_all):
    df1 = pd.DataFrame([])
    df1['salesman_id'] = df_all[["salesman_id_before"]]
    df1['type'] = 'Antes'

    df2 = pd.DataFrame([])
    df2['salesman_id'] =  df_all[["salesman_id_after"]]
    df2['type'] = 'Depois'

    df1 = df1.append(df2).reindex()

    g1 = sns.countplot(x="salesman_id", hue='type', data=df1)
    sns.despine()
    plt.show()

plot_count_clients(df_all)
```

### Faturamento

```python
df_before = pd.read_csv(_path+'original_benefit_per_salesman.csv')
df_after  = pd.read_csv(_path+'optimized_benefit_per_salesman.csv')
df = pd.merge(df_before, df_after, on='salesman_id', 
              suffixes=('_before', '_after'))[['salesman_id', 'value_before', 'value_after']]

df['salesman_id'] = df.salesman_id.apply(lambda x: str(int(x)))
df['diff'] = df.value_after-df.value_before
df = df.sort_values(['diff'], ascending=True)
df.head()
```

```python
def plt_barplot(df, title, ylabel):
    df1 = df[["salesman_id", "value_before"]]
    df1.columns = ["salesman_id", "value"]
    df1['type'] = 'Antes'

    df2 = df[["salesman_id", "value_after"]]
    df2.columns = ["salesman_id", "value"]
    df2['type'] = 'Depois'

    df = df1.append(df2).reindex()    
    # Initialize the figure with a logarithmic x axis
    #f, ax = plt.subplots(figsize=(25, 10))

    plt.title(title, loc='left', y=1.02)
    sns.barplot(y="value", x='salesman_id', orient='V', hue="type",
                color=sns.color_palette("Blues_d", n_colors=10)[0], data=df)
    
    sns.despine()
    plt.ylabel(ylabel);
    plt.xlabel('Representante');
    
    plt.show()

plt_barplot(df, 'Faturamento por representante', 'Faturamento')
```

#### Sustentabilidade

```python
df_before = pd.read_csv(_path+'original_sustainability_per_salesman.csv')
df_after  = pd.read_csv(_path+'optimized_sustainability_per_salesman.csv')
df = pd.merge(df_before, df_after, on='salesman_id', 
              suffixes=('_before', '_after'))[['salesman_id', 'value_before', 'value_after']]

df['salesman_id'] = df.salesman_id.apply(lambda x: str(int(x)))
df['diff'] = ((df.value_after-df.value_before)/np.fabs(df.value_before))*100
df = df.sort_values(['diff'], ascending=True)
df.head()
```

```python
plt_barplot(df, 'Sustentabilidade', 'Distncia')
```
